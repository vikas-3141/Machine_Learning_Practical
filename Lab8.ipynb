# Do all pre processing on diabetes csv

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib as plt
from sklearn.model_selection import KFold
```

```python
data=pd.read_csv('/content/diabetes.csv')
data
```

# Import diabetes.csv data ser using sklearn

# Check the distribution of the target

```python
data.info()
```

# Replace zeros with NaN for specific columns

```python
data.replace(0, np.nan, inplace=True)
```

# Check for missing values

```python
data.isnull().sum()
```

# Fill missing values with median

```python
data['Pregnancies'] = data['Pregnancies'].fillna(data['Pregnancies'].median())
data['Glucose'] = data['Glucose'].fillna(data['Glucose'].median())
data['BloodPressure'] = data['BloodPressure'].fillna(data['BloodPressure'].median())
data['SkinThickness'] = data['SkinThickness'].fillna(data['SkinThickness'].median())
data['Insulin'] = data['Insulin'].fillna(data['Insulin'].median())
data['BMI'] = data['BMI'].fillna(data['BMI'].median())
```

# Convert Targer data into interger code

```python
data['Outcome']=data['Outcome'].replace('Diabetic',1)
data['Outcome']=data['Outcome'].replace('Non Diabetic',0)
```

# Divide the data into input and output

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x,y
```

# import KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

# Convert all features into Z score

```python
scaller=StandardScaler()
x_s=scaller.fit_transform(x)
```

```python
x_s,len(x_s)
```

# Splitting the dataset into the Training set and Test set

```python
x_train,x_test,y_train,y_test=train_test_split(x_s,y,test_size=0.3,random_state=25)
```

# Create KNN Model

```python
model=KNeighborsClassifier()
```

# Fitting KNN   on dataset

```python
model.fit(x_train,y_train)
```

# Display classification_report

```python
KNeighborsClassifier(n_neighbors=20)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

# Use Cross Validation to find best hyperparameter for KNN

```python
print(classification_report(y_test,model.predict(x_test)))
```

```python
cross_val_score(KNeighborsClassifier(),x_train,y_train,cv=10)
```
