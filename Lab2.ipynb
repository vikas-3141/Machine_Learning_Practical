## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

## Step 2. Import the dataset

```python
data=pd.read_csv('/content/iphones.csv')
data
```

## Step 3. Import LinearRegression model from linear_model family

```python
data['iphonenumber']
```

## Step 4. Create LinearRegression Model

```python
lr=LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
a=data['iphonenumber'].values.reshape(-1,1)
b=data['price']
```

## Step 6. Fit the Model (Training)

```python
lr.fit(a,b)
```

## Step 7. Prdict the value

```python
lr.predict([[12]])
predict=lr.predict(a)
predict
```

## Step 8. Result Visualization

```python
plt.scatter(a,b)
plt.plot(a,predict)
plt.plot(a,b)
plt.xlabel('iphonenumber')
plt.ylabel('price')
plt.show()
```

## Dataset - 2 - Salary Dataset

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

## Step 2.  Read the Data

```python
Salary_Data=pd.read_csv('/content/Salary_Data.csv')
Salary_Data
```

## Step 3. View Data using Scatter plot

```python
plt.scatter(Salary_Data['YearsExperience'],Salary_Data['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

## Step 4. Splitting dataset in to input and output

```python
x=Salary_Data['YearsExperience'].values.reshape(-1,1)
y=Salary_Data['Salary']
```

## Step 4. Splitting dataset in to input and output

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
# x=Salary_Data.iloc[:,:-1].values
# y=Salary_Data.iloc[:,1].values
```

## Step 6. Create LinearRegression Model

```python
lra=LinearRegression()
```

## Step 7. Fit the Model (Training)

```python
lra.fit(X_train,Y_train)
```

##  Step 6. Predicting the Test set results

```python
y_pred=lra.predict(X_test[[1]])
accuracy=lra.score(X_train,Y_train)
y_pred
accuracy
```

##  Step 7. Visualising the Training set results

```python
plt.scatter(X_train,Y_train,color='red')
plt.plot(X_train,lra.predict(X_train),color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

##  Step 8. Visualising the Test set results

```python
plt.scatter(X_test,Y_test,color='red')
plt.plot(X_train,lra.predict(X_train),color='blue')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```
